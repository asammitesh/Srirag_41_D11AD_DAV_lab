```r
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Matplotlib Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```r
import matplotlib.pyplot as plt


# Scatter plot
plt.scatter([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```r
# Bar plot
plt.bar(['A', 'B', 'C', 'D'], [10, 20, 15, 25])
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```r
# Histogram
import numpy as np
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```r
import seaborn as sns
import plotly.express as px
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips,errorbar=None)
plt.title("Seaborn Example")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```r
fig = px.scatter(x=[1, 2, 3, 4], y=[10, 11, 12, 13])
fig.update_layout(title="Plotly Example", xaxis_title="X-axis", yaxis_title="Y-axis")
fig.show()
```

# **R**

```r
library(ggplot2)
ggplot(mtcars, aes(x = mpg, y = disp)) +
  geom_point(color = "#AA3333") +
  labs(title = "Scatter Plot of MPG vs Displacement",
       x = "MPG",
       y = "Displacement") +
  theme_minimal() +
    theme(
    text = element_text(size = 16),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    plot.title = element_text(size = 20)
  )
```

```r
library(ggplot2)

# Create sample data
data <- data.frame(
  x = c(1, 2, 3, 4),
  y = c(1, 4, 9, 16)
)

# Create ggplot object and specify aesthetics
ggplot(data, aes(x = x, y = y)) +
  # Add a line plot layer
  geom_line() +
  # Add title and axis labels
  labs(title = "Simple Line Plot Example", x = "X-axis", y = "Y-axis")
```

```r
ggplot(diamonds, aes(x = cut, fill = cut)) +
  geom_bar() +
  labs(title = "Bar Chart of Diamond Cuts",
       x = "Cut",
       y = "Count") +
  theme_minimal() +
    theme(
    text = element_text(size = 16),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    plot.title = element_text(size = 20)
  )
```

```r
library(lattice)

x <- rnorm(100)
y <- rnorm(100)

xyplot(y ~ x,
       main = "Scatter Plot",
       xlab = "X-axis",
       ylab = "Y-axis",
       pch = 16,
       col = "red",
       xlim = c(-3, 3),
       ylim = c(-3, 3)
)
```

```r
data <- rnorm(100)

histogram(~ data,
          main = "Histogram of Sample Data",
          xlab = "Value",
          ylab = "Frequency",
          col = "skyblue",
          border = "black",
          breaks = 10
)
```
